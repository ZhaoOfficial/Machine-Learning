```python
import torch
import torch.nn as nn
from torch.autograd import Variable
from torch.utils.data import DataLoader
import torchvision
import matplotlib.pyplot as plt
from matplotlib import cm
from sklearn.manifold import TSNE

torch.manual_seed(1)
```

## Recipe 3-6. Implementing a Convolutional Neural Network (CNN)

```python
# hyper-parameters
EPOCHES = 1000
# the nubmer of samples at a time to pass the epoch
BATCH_SIZE = 50
# learning rate
LEARNING_RATE = 0.001
# download mnist
mnist_data = torchvision.datasets.MNIST(
    root='data/',
    train=True,
    transform=torchvision.transforms.ToTensor()
)

print(mnist_data.data.shape)
print(mnist_data.targets.shape)
print(mnist_data.data.dtype)
plt.imshow(mnist_data.data[0], cmap='gray')
plt.axis('off')
plt.show()
```

```python
mnist_training_loader = DataLoader(
    dataset=mnist_data,
    batch_size=BATCH_SIZE,
    shuffle=True
)
mnist_testing_data = torchvision.datasets.MNIST(
    root='data/',
    train=False
)
mnist_testing_x = Variable(
    torch.unsqueeze(mnist_testing_data.data, dim=1)
).type(torch.FloatTensor)
mnist_testing_y = mnist_testing_data.targets
print(mnist_testing_x.shape)
print(mnist_testing_y.shape)
print(mnist_testing_x.dtype)
```

```python
# Convolutional Neural Networks
class CNN(nn.Module):
    def __init__(self):
        super(CNN, self).__init__()
        self.conv1 = nn.Sequential(     # input.shape = (1, 28, 28)
            nn.Conv2d(
                in_channels=1,          # input height
                out_channels=16,        # n filters
                kernel_size=5,          # filter.shape = (5, 5)
                stride=1,               # filter step
                padding=2
            ),                          # output.shape = (16, 28, 28)
            nn.ReLU(),                  # activations
            nn.MaxPool2d(kernel_size=2) # output.shape = (16, 14, 14)
        )
        self.conv2 = nn.Sequential(     # input.shape = (16, 14, 14)
            nn.Conv2d(16, 32, 5, 1, 2), # output.shape = (32, 14, 14)
            nn.ReLU(),
            nn.MaxPool2d(2)
        )                               # output.shape = (32, 7, 7)
        self.out = nn.Linear(32 * 7 * 7, 10) # fully connected layer
    
    def forward(self, image):
        x = self.conv1(image)
        x = self.conv2(x)
        x = x.view(x.shape[0], -1)
        output = self.out(x)
        return output, x
```

```python
cnn = CNN()
print(cnn)
optimizer = torch.optim.Adam(cnn.parameters(), lr=LEARNING_RATE)
loss_fn = nn.CrossEntropyLoss()
```

```python
def plot_with_labels(lowDWeights, labels):
    plt.cla()
    X, Y = lowDWeights[:, 0], lowDWeights[:, 1]
    for x, y, s in zip(X, Y, labels):
        c = cm.rainbow(255 * s // 9)
        plt.text(x, y, s, backgroundcolor=c, fontsize=9)
    plt.xlim(X.min(), X.max())
    plt.ylim(Y.min(), Y.max())
    plt.title("Visualize last layer")
    plt.pause(1)
plt.ion()
```

```python
for epoch in range(10):
    for step, (x, y) in enumerate(mnist_training_loader):
        # batch_x.shape = (BATCH_SIZE, 1, row, col)
        # batch_x.dtype = float
        batch_x = Variable(x)
        # batch_y.shape = (BATCH_SIZE)
        batch_y = Variable(y)
        
        # output.shape = (BATCH_SIZE, 10)
        output, _ = cnn(batch_x)
        loss = loss_fn(output, batch_y)
        optimizer.zero_grad()
        loss.backward()
        optimizer.step()

        if step % 100 == 0 and epoch == 0:
            test_y_pred, last_layer = cnn(mnist_testing_x)
            test_y_pred = torch.max(test_y_pred, dim=1).indices
            accuracy = (test_y_pred == mnist_testing_y).sum()
            print('Epoch: {} | step: {} | train loss = {:.4f} | test accuracy = {} / {}'.
                  format(epoch, step, loss.data, accuracy, mnist_testing_y.shape[0]))

            tsne = TSNE(perplexity=30, n_components=2, init='pca', n_iter=5000)
            plot_only = 500
            low_dim_embs = tsne.fit_transform(last_layer.data.numpy()[:plot_only, :])
            labels = mnist_testing_y.numpy()[:plot_only]
            plot_with_labels(low_dim_embs, labels)
            
    test_y_pred, last_layer = cnn(mnist_testing_x)
    test_y_pred = torch.max(test_y_pred, dim=1).indices
    accuracy = (test_y_pred == mnist_testing_y).sum()
    print('Epoch: {} | train loss = {:.4f} | test accuracy = {} / {}'.
          format(epoch, loss.data, accuracy, mnist_testing_y.shape[0]))

            
plt.ioff()
```

## Recipe 3-7. Reloading a Model
Store and re-upload a model that has already been trained.

```python
# x.shape = (100, 1)
x = torch.unsqueeze(torch.linspace(-1, 1, 100), dim=1)
# noisy data
y = x.pow(2) + 0.2 * torch.rand(x.shape)
x, y = Variable(x, requires_grad=False), Variable(y, requires_grad=False)
```

```python
def save():
    sample_net = torch.nn.Sequential(
        torch.nn.Linear(1, 10),
        torch.nn.ReLU(),
        torch.nn.Linear(10, 1)
    )
    optimizer = torch.optim.SGD(sample_net.parameters(), lr=0.1)
    loss_fn = torch.nn.MSELoss()
    
    for epoch in range(100):
        prediction = sample_net(x)
        loss = loss_fn(prediction, y)
        optimizer.zero_grad()
        loss.backward()
        optimizer.step()
    
    plt.figure(1, figsize=(10, 3))
    plt.subplot(1, 3, 1)
    plt.title('sample_net')
    plt.scatter(x.data.detach().numpy(), y.data.detach().numpy())
    plt.plot(x.data.detach().numpy(),
             prediction.data.detach().numpy(), 'r-', lw=5)
    plt.legend(['x-prediction', 'x-y'])
    
    # 2 ways to save the net
    # 1. save the entire net
    torch.save(sample_net, 'data/sample_net.pkl')
    # 2. save the parameters
    torch.save(sample_net.state_dict(), 'data/sample_net_params.pkl')
```

```python
def load_net():
    loaded_net = torch.load('data/sample_net.pkl')
    prediction = loaded_net(x)
    plt.subplot(1, 3, 2)
    plt.title('load_net')
    plt.scatter(x.data.detach().numpy(), y.data.detach().numpy())
    plt.plot(x.data.detach().numpy(),
             prediction.data.detach().numpy(), 'r-', lw=5)
    plt.legend(['x-prediction', 'x-y'])
```

```python
def load_params():
    loaded_net = torch.nn.Sequential(
        torch.nn.Linear(1, 10),
        torch.nn.ReLU(),
        torch.nn.Linear(10, 1)
    )
    loaded_net.load_state_dict(torch.load('data/sample_net_params.pkl'))
    prediction = loaded_net(x)
    plt.subplot(1, 3, 3)
    plt.title('load_params')
    plt.scatter(x.data.detach().numpy(), y.data.detach().numpy())
    plt.plot(x.data.detach().numpy(),
             prediction.data.detach().numpy(), 'r-', lw=5)
    plt.legend(['x-prediction', 'x-y'])
```

```python
save()
load_net()
load_params()
plt.show()
```
